```python
#1.4-Una variable numèrica i una categòrica (Districte i Preu)

df = pd.read_csv("preu.csv")                                            # Carrego les dades
df['Preu'] = df['Preu'].replace('--',"NAN")                             # Reemplaço els valors nuls per NAN a la columna preu
df['Preu']=df['Preu'].astype(float)                                     # Un cop no hi han valors nuls, defineixo els valors de la columna com a números. 

df = df.query("Lloguer_mitja=='Lloguer mitjà mensual (Euros/mes)'")     # Faig una consulta per a que a la columna preu aparegui només el lloguer mitjà mensual
df = df.groupby ("Nom_Districte").mean()
my_plot_1v4 = df.plot.bar(y='Preu')
```

```python
#1.5-Dues variables numèriques (Any o Trimestre i Preu). Primer intent
df = pd.read_csv("preu.csv")
df['Preu'] = df['Preu'].replace('--',"NAN")
df['Preu']=df['Preu'].astype(float)

dh = df.query("Lloguer_mitja=='Lloguer mitjà mensual (Euros/mes)'")

dt = dh.plot( 1,7, kind="scatter")
```

```python
#1.5.2-Dues variables numèriques (Any o Trimestre i Preu). Segon intent

print('Dues variables numèriques (Any o Trimestre i Preu)')

df = pd.read_csv("preu.csv")                                            # Carrego les dades
df['Preu'] = df['Preu'].replace('--',"NAN")                             # Reemplaço els valors nuls per NAN a la columna preu
df['Preu']=df['Preu'].astype(float)                                     # Un cop no hi han valors nuls, defineixo els valors de la columna com a números. 

df = df.query("Lloguer_mitja=='Lloguer mitjà mensual (Euros/mes)'")     # Faig una consulta per a que a la columna preu aparegui només el lloguer mitjà mensual

df_promedio = df.groupby("Trimestre", as_index=False).mean()            # Agrupo els valors per trimestre i faig la mitja dels valors de la resta de columnes. 


print(dt.set_title("Preu mig per trimestre"))                           # Poso un titol
my_plot_1v5 = df_promedio.plot( 0,4, kind="bar")                        # Defineixo el gràfic
my_plot_1v5.set_ylim(800, 900)                                          # Defineixo que l'escala de l'eix Y comenci en 800 i acabi en 900 
my_plot_1v5.grid()                                                      # Defineixo que aparegui una graella

```

```python
#1.6-Tres variables (Barri o Districte, Trimestre i Preu)

print('Tres variables (Barri o Districte, Trimestre i Preu)')


df = pd.read_csv("preu.csv")                                            # Carrego les dades
df['Preu'] = df['Preu'].replace('--',"NAN")                             # Reemplaço els valors nuls per NAN a la columna preu
df['Preu']=df['Preu'].astype(float)                                     # Un cop no hi han valors nuls, defineixo els valors de la columna com a números. 

df = df.query("Lloguer_mitja=='Lloguer mitjà mensual (Euros/mes)'")     # Faig una consulta per a que a la columna preu aparegui només el lloguer mitjà mensual

my_plot_1v6 = plt.scatter(df['Nom_Districte'],df['Preu'], df['Trimestre'], color='grey')

```
